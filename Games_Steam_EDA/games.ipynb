```python
df_req = pd.read_csv('./output.csv', sep=',')
df_req.head()
```

#### Data Analysis

```python
null_counts = df_req.isnull().sum()
print(null_counts)
# wen have just objects data type
# we will drop nulls to clean the data
```

```python
df_req.dropna(inplace=True)
df_req.info()
```

```python
# we will separate the columns that contains numeric datatypes and strings
df_req['Memory_int'] = df_req['Memory:'].str.extract('(\d+)').astype(float)
df_req['Memory_str'] = df_req['Memory:'].str.extract('(\w+\S)').astype(str)
df_req['FileSize_int'] = df_req['File Size:'].str.extract('(\d+)').astype(float)
df_req['FileSize_str'] = df_req['File Size:'].str.extract('(\s\D+)').astype(str)
df_req = df_req.drop(['Memory:', 'File Size:'], axis=1)

df_req.loc[df_req['name'].str.contains('System Requirements'), 'name'] = df_req['name'].str.replace('System Requirements', '')

df_req.head(4)
```

```python
df_req['Graphic_card_brand'] = 'Other'
df_req['Graphic_card_brand'] = np.where(df_req['Graphics Card:'].str.contains('NVIDIA'), 'NVIDIA', df_req['Graphic_card_brand'])
df_req['Graphic_card_brand'] = np.where(df_req['Graphics Card:'].str.contains('AMD'), 'AMD', df_req['Graphic_card_brand'])
df_req['Graphic_card_brand'] = np.where(df_req['Graphics Card:'].str.contains('ATI'), 'AMD', df_req['Graphic_card_brand'])
df_req['Graphic_card_brand'] = np.where(df_req['Graphics Card:'].str.contains('Intel'), 'Intel', df_req['Graphic_card_brand'])
df_req.head()
```

### Save the file cleaned ready to PBI

```python
df_req.to_csv('./Game_requirements_Clean.csv')
```
