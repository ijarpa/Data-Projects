```python
import pandas as pd
```

```python
df = pd.read_csv("./mobiles.csv", encoding='utf-8')
df.head()

# price in indian rupias (₹)
# need to clean the price
```

```python
# Transforming columns
df['price_details'] = df["price_details"].str.extract(r'([\d,]+)')[0] #cleaning the price
df[['RAM','ROM',"Memory_Expandable"]] = df["memory"].str.split('|', expand=True) #separating the memory details
df[['Camera','Front_Camera',]] = df["camara_info"].str.split('|', expand=True) #separating the memory details
df[['display_cm', 'display_inch', 'resolution']] = df['display'].str.extract(r'(.+)\s\((.+)\)(.+)') #separating the display information
df[['battery_mAh', 'battery_description']] = df['battery'].str.split(' ', 1, expand=True) #
```

```python
df.drop(['battery', 'display', 'camara_info'], axis=1, inplace=True)
```
