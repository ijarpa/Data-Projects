```python
import pandas as pd
import numpy as np
import re
```

```python
bancos = [
    'b1202304001',
    'b1202304037',
    'b1202304016',
    'b1202304012',
    'b1202304014'
]
```

```python
dfs = {}  # Diccionario para almacenar los DataFrames

for banco in bancos:
    ruta = f'./PruebaPM/{banco}.txt'
    df = pd.read_csv(ruta, sep='\t', encoding='latin1', header=None)
    df['Archivo'] = banco
    
    dfs[banco] = df 


df_final = pd.concat(dfs.values(), ignore_index=True)
```

```python
df_names = {0:'Cuenta_contable', 1:'Saldo_CLP',2:'Saldo_UF',3:'Saldo_USD',4:'Saldo_otras_monedas', 'Archivo':'Archivo'}
df_final.rename(columns=df_names, inplace=True)
df_final
```

```python
df_final = df_final.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df_final.iloc[:, :4] = df_final.iloc[:, :4].replace(',', '.', regex=True).astype(float)
df_final.iloc[:, :4] = df_final.iloc[:, :4].astype(float)
```
