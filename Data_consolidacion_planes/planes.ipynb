```python
import pandas as pd
import numpy as np
import os
```

```python
path = 'files'
files_names = os.listdir(path)
```

```python
dfs = {} # dataframes dictionary

for file in files_names:
    df = pd.read_excel(f'./files/{file}')

    df['Proyecto'] = df.iloc[0,0] # crear una columna con el nombre del proyecto
    df = df.iloc[19:, [0,1,2,-1]] # eliminar las primeras filas y las columnas vacias
    df = df.drop(df.columns[1], axis=1) # elimina la columna intermedia con datos vacios
    df = df.dropna(axis=0, how='any') # quitar las filas que tengan datos vacios

    dfs[file] = df # Agregar el dataframe al diccionario

df_concat = pd.concat(dfs.values(), ignore_index=True) # Concatenar los dataframes
```

```python
column_names = {'Unnamed: 0': 'cuenta', 'Unnamed: 2': 'Nombre_cuenta', 'Proyecto':'Proyecto'}
df_concat.rename(columns=column_names, inplace=True)
df_concat
```

```python
df_concat.to_csv('all_data.csv', index=False)
```
